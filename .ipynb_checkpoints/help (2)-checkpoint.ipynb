```python
df.head()
df.drop(columns=['Rating text'],inplace=True)
```

```python
cuisines = df['Cuisines'].str.get_dummies(sep=', ')
df_encoded = df.join(cuisines)
```

```python
df_encoded
```

```python
features = df_encoded.drop(['Restaurant Name', 'Cuisines'], axis=1)
```

```python
similarity_matrix = cosine_similarity(features)
np.fill_diagonal(similarity_matrix, 0)  # Ignore self-similarity

# Recommendation Function
def recommend_restaurant(restaurant_index, num_recommendations=15):
    similarities = similarity_matrix[restaurant_index]
    similar_restaurants = similarities.argsort()[-num_recommendations:][::-1]
    return df.iloc[similar_restaurants]

# Example Recommendation
print(recommend_restaurant(0))
```
