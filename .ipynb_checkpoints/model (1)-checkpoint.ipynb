```python
# creating a new column for food types and fill in with nan values 
df['food types'] = np.nan
df['food types'] = df['food types'].astype('str')
```

```python
def classify_food_type(row):
    ingredients = row['ingredients']
    calories = row['calories']

    # Check for Veg dessert and Non-Veg dessert
    if 'eggs' not in ingredients:
        if any(item in ingredients for item in ['ice-cream', 'chocolate', 'cookies']):
            return 'Veg dessert'
    elif 'eggs' in ingredients:
        if any(item in ingredients for item in ['ice-cream', 'chocolate', 'cookies']):
            return 'Non-Veg dessert'

    # Check for Healthy type
    if pd.notna(calories) and 20 < calories < 300:
        return 'Healthy'
    
    # Check for Non-veg type
    if any(item in ingredients for item in ['chicken', 'eggs', 'ham', 'pepperoni']):
        return 'Non-veg'

    # Default to Veg
    return 'Veg'

# Apply the function to classify food types
df['food types'] = df.apply(classify_food_type, axis=1)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# # Vectorize the ingredients
# tfidf = TfidfVectorizer(stop_words='english')
# tfidf_matrix = tfidf.fit_transform(df['ingredients_str'])

# Compute cosine similarity
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)

# Function to recommend restaurants
def recommend_restaurants(restaurant_name, cosine_sim=cosine_sim):
    idx = df.index[df['recipe_name'] == restaurant_name][0]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:11]  # Get top 10 recommendations
    restaurant_indices = [i[0] for i in sim_scores]
    return df['restaurant'].iloc[restaurant_indices]

# Example recommendation
print(recommend_restaurants("amish tomato ketchup for canning"))
```

```python
from sklearn.decomposition import TruncatedSVD
svd = TruncatedSVD(n_components=50)
reduced_features = svd.fit_transform(tfidf_matrix)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cosine_sim = cosine_similarity(reduced_features)

# Function to recommend restaurants
def recommend_restaurants(restaurant_name, df, cosine_sim):
    idx = df.index[df['recipe_name'] == restaurant_name][0]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:11]  # Top 10 recommendations
    restaurant_indices = [i[0] for i in sim_scores]
    return df['restaurant'].iloc[restaurant_indices]

# Example recommendation
recommendations = recommend_restaurants("amish tomato ketchup for canning", df, cosine_sim)
print(recommendations)
```

```python
import faiss
import numpy as np

# Normalize your features to make it compatible with cosine similarity
faiss.normalize_L2(reduced_features)

# Build the index for nearest neighbor search
index = faiss.IndexFlatIP(reduced_features.shape[1])
index.add(reduced_features)

# Perform the search
D, I = index.search(reduced_features, 10)  # Top 10 similar items for each item

# Accessing the results
for i, (distances, indices) in enumerate(zip(D, I)):
    print(f"Top 10 recommendations for index {i}: {indices}")
```
